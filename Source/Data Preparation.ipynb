```python
# Importing all the important libraries
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading all the excel tables

city_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/City.xlsx")
continent_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/Continent.xlsx")
country_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/Country.xlsx")
item_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/Item.xlsx")
mode_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/Mode.xlsx")
region_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/Region.xlsx")
transaction_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/Transaction.xlsx")
type_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/Type.xlsx")
user_df = pd.read_excel("C:/Users/Saima Modak/Capstone Projects/Tourism Analysis/Datasets/User.xlsx")
```

```python
# Checking for null values
for x, df in zip(
["City", "Continent", "Country", "Item", "Mode", "Region", "Transaction", "Type",  "User"], 
[city_df, continent_df, country_df, item_df, mode_df, region_df, transaction_df, type_df, user_df]
):

    print(f"{x}:\n{df.isnull().any()}")    
```

```python
# Removing the row with null value
city_df = city_df.dropna()
```

```python
# Getting the column names of all the tables
for x, df in zip(
["City", "Continent", "Country", "Item", "Mode", "Region", "Transaction", "Type", "User"], 
[city_df, continent_df, country_df, item_df, mode_df, region_df, transaction_df, type_df, user_df]
):
    print(f"{x}: {df.columns}")
```

```python
# Renaming the column names for required tables
region_df.rename(columns={'ContentId': 'ContinentID'}, inplace=True)
country_df.rename(columns={'RegionId': 'Country_RegionId'}, inplace=True)
user_df.rename(columns={'CityId': 'User_City_Id', "ContenentId": "ContinentID"}, inplace=True)
continent_df.rename(columns = {"ContenentId": "ContinentID", "Contenent": "Continent"}, inplace=True)
for x , df in zip(
    ["Region", "Country", "User", "Continent"],[region_df, country_df, user_df, continent_df]
):
    print(f"{x}:\n{df.columns}")
```

```python
# Merging the tables to get a single dataset
merged_df = transaction_df.copy()

# Merging user data with transaction data
merged_df = merged_df.merge(user_df, on = "UserId", how = "inner")

# Merging with Item on AttractionId
merged_df = merged_df.merge(item_df, on = "AttractionId", how = "inner")

# Mergeing with Mode on VisitMode
merged_df = merged_df.merge(mode_df, left_on = 'VisitMode', right_on = 'VisitModeId', how ='inner')

# Merging with Type on AttractionTypeId (from Item)
merged_df = merged_df.merge(type_df, on = "AttractionTypeId", how = "inner")

# Merging with Continent on ContinentID (from User)
merged_df = merged_df.merge(continent_df, on = "ContinentID", how = "inner")

# Merging with Region on RegionId (from User)
merged_df = merged_df.merge(region_df, on = "RegionId", how = "inner")

# Merging with Country on CountryId (from User)
merged_df = merged_df.merge(country_df, on ='CountryId', how = "inner")

# Merging with City on AttractionCityId from Item to CityId
df = merged_df.merge(city_df, left_on ='AttractionCityId', right_on ='CityId', how = "inner")
```

#### Data Cleaning and Preprocessing

```python
df.head(2)
```

```python
# Dropping duplicate columns
df = df.drop(['VisitMode_x', 'ContinentID_x', 'AttractionCityId'], axis=1)

# Renaming the country ID columns
df = df.rename(columns={
    'CountryId_x': 'UserCountryId',      # Country of the user
    'CountryId_y': 'AttractionCountryId',  # Country of the attraction
    'VisitMode_y': 'VisitMode',
    'ContinentID_y' : 'ContinentID'
    })
```

```python
df.head()
```

```python

df.isna().sum()
```

```python
# Filling missing User_City_ID with a default value (like -1)
df['User_City_Id'] = df['User_City_Id'].fillna(-1)

# Adding a flag column to indicate originally missing data
df['Missing_City_ID'] = df['User_City_Id'] == -1
```

```python
# Standardizing city names (removing extra spaces, standardizing case)
df['CityName'] = df['CityName'].str.strip().str.title()
```

```python
# Resolving categorical variable discrepancies
categorical_cols = ['VisitMode', 'AttractionType', 'Continent', 'Region', 'Country']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].str.strip()
```

```python
# Removing invalid ratings
print(f"Invalid ratings (outside 1-5): {len(df[(df['Rating'] < 1) | (df['Rating'] > 5)])}")
df = df[(df['Rating'] >= 1) & (df['Rating'] <= 5)]
```

```python
# Handle outliers in Rating
Q1 = df['Rating'].quantile(0.25)
Q3 = df['Rating'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"\nRating outliers: {len(df[(df['Rating'] < lower_bound) | (df['Rating'] > upper_bound)])} values")

# Capping outliers
df['Rating'] = df['Rating'].clip(lower=1, upper=5)  # Assuming ratings are 1-5 scale
```

#### Feature Engineering

```python
# Time-based features
df['VisitSeason'] = df['VisitMonth'].map({
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
})
df['VisitQuarter'] = ((df['VisitMonth'] - 1) // 3) + 1
```

```python
## Sorting by time to ensure no leakage
df = df.sort_values(['UserId', 'VisitYear', 'VisitMonth'])

# Calculating user features based on previous behavior only
df['user_previous_visits'] = df.groupby('UserId').cumcount()
df['user_avg_rating_before'] = df.groupby('UserId')['Rating'].transform(lambda x: x.shift(1).expanding().mean())
df['user_avg_rating_before'] = df['user_avg_rating_before'].fillna(3.0)  # neutral default
df['user_rating_trend'] = df.groupby('UserId')['Rating'].diff()
```

```python
# Calculating attraction features based on previous behavior only
def get_previous_stats(group):
    # Calculating previous visits for each attraction
    group = group.reset_index(drop=True)
    group['attraction_previous_visits'] = range(len(group))
    # Calculating rolling average rating excluding current visit  
    group['attraction_avg_rating_before'] = group['Rating'].shift(1).expanding().mean()
    group['attraction_avg_rating_before'] = group['attraction_avg_rating_before'].fillna(3.0)
    # Counting unique previous visitors
    group['attraction_previous_visitors'] = group.groupby('UserId').cumcount().shift(1).max()
    group['attraction_previous_visitors'] = group['attraction_previous_visitors'].fillna(0)
    return group

df = df.groupby('AttractionId').apply(get_previous_stats).reset_index(drop=True)
```

```python
# City-level features (aggregate stats that don't use target)
city_stats = df.groupby('User_City_Id').size().to_frame('city_popularity')
df = df.merge(city_stats, left_on='User_City_Id', right_index=True, how='left')
df['city_popularity'] = df['city_popularity'].fillna(0)
```

```python
# Creating interaction features
df['user_continent'] = df['UserId'].astype(str) + '_' + df['Continent']
df['user_attraction_type'] = df['UserId'].astype(str) + '_' + df['AttractionTypeId'].astype(str)
df['attraction_type_season'] = df['AttractionTypeId'].astype(str) + '_' + df['VisitSeason']
```
